```python
# Step 1: Install Tesseract
!apt install -y tesseract-ocr
!pip install pytesseract

# Step 2: Import the required libraries
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt
import re

# Step 3: Load an image
img_path = '/content/dataset/test/images/IMG_20241020_213142_211_jpg.rf.d66311771b18c868cf1cb7559eb8112e.jpg'
img = Image.open(img_path)

# Step 4: Convert the image to grayscale
img_gray = img.convert('L')  # Convert to grayscale

# Step 5: Display the grayscale image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()

# Step 6: Run OCR on the grayscale image
text = pytesseract.image_to_string(img_gray)

# Step 7: Print the extracted text
print("Extracted Text:")
print(text)

# Step 8: Extract the expiry date using regex
expiry_date_pattern = r'\b\d{1,2}[-/]\d{1,2}[-/]\d{2,4}\b'
expiry_dates = re.findall(expiry_date_pattern, text)


if expiry_dates:
    print("Extracted Expiry Dates:")
    for date in expiry_dates:
        print(date)
else:
    print("No expiry date found.")
```
